```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_olimpic=pd.read_csv("https://raw.githubusercontent.com/lihkirun/AppliedStatisticMS/main/DataVisualizationRPython/Lectures/Python/PythonDataSets/athlete_events.csv")
data_olimpic_2016=data_olimpic.loc[data_olimpic['Year']==2016]
final_2016=data_olimpic_2016.loc[data_olimpic_2016["Medal"].notna()]
```

Estamos interezados en los ganadores de medallas en los deportes de atletismo, natación, remo, fútbol y hockey en el año 2016, que según vimos en el resultado anterior obtivimos estos datos en el DataFrame dep5_2016.

```python
def five_import(x):
    return x in ["Rowing", "Athletics", "Swimming", "Football", "Hockey"]
    

dep5_2016=final_2016.loc[final_2016["Sport"].apply(five_import)]
dep5_2016.head()
```

1. Crea un gráfico utilizando una técnica de visualización adecuada que presente de la mejor manera posible el patrón global de las características de height y weight de los ganadores de medallas de 2016 de los cinco deportes.

```python
sns.set_theme(style="ticks")
sns.jointplot(data=dep5_2016, y="Weight", x="Height", kind="hex", color="#5E548E")
plt.show()
```

**Conclusion:** El gráfico anterior podemos observar que existe una aparente relación positiva entre el peso y la estatura de los campeones olimpicos en el 2016. Existe la tendencia de que a medida que aumenta la estatura se espera que el peso aumente.

2. Crea un gráfico utilizando una técnica de visualización adecuada que presente de la mejor manera posible la estadística de resumen para la altura y el peso de los jugadores que ganaron cada tipo de medalla (oro/plata/bronce) en los datos.

```python
colors = ["#5E548E",   "#9F86C0",  "#BE95C4" ]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

sns.boxplot(y="Height", data=dep5_2016, hue='Medal', palette=colors,ax=axes[0])
axes[0].set_title('Altura de los medallistas olímpicos del 2016')

sns.boxplot(y="Weight", data=dep5_2016, hue='Medal', palette=colors,ax=axes[1])
axes[1].set_title('Pesos de los medallistas olímpicos del 2016')

plt.show()
```

**Conclusión:** En la gráfica anterior, podemos observas mediante los dos diagramas de cajas diferentes estadísticos de las varibles Peso y Altura. Del primer diagrama podemos observar que la media de estatura en los tres grupos de la categoría medallas son muy simirales y es cercana a 178 cm, al parecer en los tres grupos no hay mucha variabilidad en los datos puesto que la extensión de als cajas no es muy grande, además podemos resaltar la precencia de un dato atípico. 

Por parte del diagrama del peso podemos ver que la media de los 3 grupos de medallas estan cerca de los 70 kilos, con mas precensia de datos atípicos y esta vez en los tres grupos, por parte de la variabilidad de los datos, vemos que es mucho menor que en la variable altura.
